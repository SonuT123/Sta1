Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Estimation statistics is a branch of statistics that focuses on estimating unknown population
     parameters (such as the population mean, variance, proportion, etc.) based on information obtained
     from a sample. Estimation allows us to make educated guesses or inferences about population parameters
     without having to collect data from the entire population.
    
There are two common methods of estimation in statistics:
    
    1.Point Estimate:
        
        A point estimate is a single value that serves as the best guess or approximation for the 
        unknown population parameter. It's calculated using a sample statistic.
        >The most commonly used point estimate is the sample mean (x̄) when estimating 
            the population mean (μ).
        >ther point estimates include the sample proportion (p̂) when estimating the population 
            proportion (p) or the sample standard deviation (s) when estimating the population
            standard deviation (σ), among others.
            
        >Point estimates are straightforward and easy to compute but provide only a single 
            value without any information about the precision of the estimate.
            
    2.Interval Estimate (Confidence Interval):
        
        >An interval estimate, often referred to as a confidence interval,
          provides a range or interval of values within which the unknown population parameter 
          is likely to fall, along with a level of confidence.
        
        >A confidence interval consists of two values: a lower bound and an upper bound. 
           It is usually expressed as (lower limit, upper limit).
            
        >The level of confidence, often denoted as (1 - α), represents the probability
           that the interval contains the true parameter. Common confidence levels include
          90%, 95%, and 99%.
        >A larger confidence level results in a wider confidence interval,
           providing greater confidence that the true parameter lies within the interval.
            
        >Confidence intervals provide a measure of uncertainty and convey both point estimation 
            and a range of plausible values for the population parameter.
            
            
        For example, if you want to estimate the average income (μ) of a certain population,
          you could take a sample, calculate the sample mean (x̄), and then create a confidence interval,
            such as "We are 95% confident that the population mean income is between $40,000 and $45,000."
            In this case, $40,000 and $45,000 are the lower and upper limits of the confidence interval.

   
            
        

Q2. Write a Python function to estimate the population mean using a sample mean and standard
     deviation.

```python
def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    standard_error = sample_std_dev / (sample_size ** 0.5)
    

    margin_of_error = 1.96 * standard_error  
    
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    
    return sample_mean, (lower_bound, upper_bound)


sample_mean = 50  
sample_std_dev = 10  
sample_size = 100  

estimated_mean, confidence_interval = estimate_population_mean(sample_mean, sample_std_dev, sample_size)

print(f"Estimated Population Mean: {estimated_mean}")
print(f"95% Confidence Interval: ({confidence_interval[0]}, {confidence_interval[1]})")
"""
In this function:

  >We first calculate the standard error, which is the standard deviation of the sample mean. 
      It's computed as the sample standard deviation divided by the square root of the sample size.
   >We then calculate the margin of error using the critical value of 1.96, which corresponds to a 
      95% confidence level for a two-tailed test.
       Finally, we calculate the lower and upper bounds of the confidence interval by subtracting
       and adding the margin of error to the sample mean, respectively.
     >Replace the sample_mean, sample_std_dev, and sample_size with your own values to estimate the 
        population mean and confidence interval for your specific dataset.
        
        """
```

Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

 Null hypothesis (H0): The average weight of male college students 
      is equal to the average weight of female college students
        
 Alternative hypothesis (Ha): The average weight of male college students is greater
       than the average weight of female college students.  
    
   This is a directional hypothesis because it states the specific direction of the expected difference between the two groups.
   
    This hypothesis can be tested using a t-test, which is a statistical test that can be used to compare the means
      of two normally distributed populations.
        
    If the t-test shows that there is a statistically significant difference between the mean weights of the two groups,
      then we can reject the null hypothesis and conclude that the average weight of male college students 
      is greater than the average weight of female college students.
    
    However, if the t-test does not show a statistically significant difference between the mean weights of the two groups,
        then we cannot reject the null hypothesis and cannot conclude that the average weight of male college students is
        greater than the average weight of female college students.   
        
        
     It is important to note that hypothesis testing does not prove or disprove anything. 
           It simply allows us to make inferences about populations based on samples.        
           

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
     given a sample from each population.

```python
import scipy.stats as stats


sample1 = [84, 88, 92, 78, 96, 90, 82, 88, 94, 86]


sample2 = [76, 80, 70, 82, 74, 88, 78, 86, 80, 82]

alpha = 0.05


t_statistic, p_value = stats.ttest_ind(sample1, sample2)

if p_value < alpha:
    print("Reject the null hypothesis.")
    print("There is enough evidence to suggest a significant difference between the two population means.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is not enough evidence to suggest a significant difference between the two population means.")

print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
"""
    >We have two samples, sample1 and sample2, which represent data from two populations.
    >We set the significance level (alpha) to 0.05, indicating a 95% confidence level.
    >We perform a two-sample t-test using stats.ttest_ind() to test whether there is a
        significant difference between the means of the two populations.
    >We compare the p-value to alpha to make a decision: If the p-value is less than alpha,
             we reject the null hypothesis in favor of the alternative hypothesis,
             indicating a significant difference between the population means. Otherwise,
             we fail to reject the null hypothesis.
    >Finally, we print the t-statistic and p-value for reference."""
```

Q7: Write down the steps involved in hypothesis testing.

1.Formulate the Null Hypothesis (H0): The null hypothesis is a statement that there is no significant effect,
    difference, or relationship between the variables being studied. It represents the status quo or a 
    default assumption. It is denoted as H0.
    
2.Formulate the Alternative Hypothesis (Ha or H1): The alternative hypothesis is a statement
     that contradicts the null hypothesis. It suggests that there is a significant effect,
     difference, or relationship between the variables. It is what you are trying to demonstrate
      or find evidence for. It is denoted as Ha or sometimes H1.
    
3.Choose the Significance Level (α): The significance level, denoted as α, represents the probability
      of making a Type I error (rejecting the null hypothesis when it is true).
      Common choices for α are 0.05 (5%) or 0.01 (1%), but it can be set based on
        the specific needs of the study.
        
4.Collect Data: Collect a sample of data relevant to your research question.
      The sample should be random and representative of the population you are interested in.
    
5.Perform Statistical Analysis: Conduct the appropriate statistical test based on the nature of
     your data and research question. Common tests include t-tests, chi-squared tests,
     ANOVA, correlation analysis, and regression analysis, among others.
        
6.Calculate the Test Statistic: The test statistic is a value computed from your sample data 
    that quantifies the difference between your sample and what would be expected under the null
    hypothesis. The choice of test statistic depends on the statistical test used.
    
7.Determine the Critical Region: Identify the critical region in the probability distribution of the
     test statistic. This is the region of extreme values that would lead to rejecting the null hypothesis
     if your calculated test statistic falls within it. The critical region is determined based on the
     significance level (α).
8.Compare the Test Statistic to Critical Values: Compare your calculated test statistic to the critical
    values (critical region boundaries) from the probability distribution. If the test statistic
    falls into the critical region, you reject the null hypothesis. If it falls outside the critical
    region, you fail to reject the null hypothesis.
    
9.Make a Decision: Based on the comparison in step 8, make a decision regarding the null hypothesis

  >If the test statistic is in the critical region: Reject the null hypothesis in favor of the
    alternative hypothesis.
    
  >If the test statistic is outside the critical region: Fail to reject the null hypothesis.
        This does not prove the null hypothesis is true; it simply means you don't have enough 
         evidence to reject it.
10.Draw a Conclusion: Interpret the results in the context of your research question and the decision
    made in step 9. State whether you found evidence to support the alternative hypothesis or not. 
    
11.Report the Results: Communicate the findings, including the test statistic, p-value (if applicable),
    and the conclusion, in a clear and concise manner.
    
12.Consider Limitations: Discuss any limitations of the study, potential sources of error,
     and implications of the findings.

Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
     parameter set to 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

degrees_of_freedom = 10


x = np.linspace(-4, 4, 1000)  
pdf = t.pdf(x, df=degrees_of_freedom)  


plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Student\'s t-distribution (df={degrees_of_freedom})', color='blue')
plt.title('Student\'s t-distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

plt.show()

"""
>We import the necessary libraries: numpy for numerical operations, matplotlib.pyplot 
  for creating plots, and scipy.stats.t for the Student's t-distribution functionality.
  
>We set the degrees of freedom to 10, as specified.
>We generate a range of x values for the plot using np.linspace.
>We calculate the probability density function (PDF) of the Student's t-distribution using t.pdf
>We create the plot, add a title, labels, a legend, and enable gridlines.
>Finally, we display the plot using plt.show()
>Running this code will generate a plot of the Student's t-distribution with 10 degrees of freedom.
  Adjust the x range or other parameters as needed to customize the plot further
```

Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
     random samples of equal size and a null hypothesis that the population means are equal.

```python
import numpy as np
from scipy import stats


np.random.seed(0)  
sample_size = 30
sample1 = np.random.normal(loc=5, scale=2, size=sample_size)
sample2 = np.random.normal(loc=5, scale=2, size=sample_size)


t_statistic, p_value = stats.ttest_ind(sample1, sample2)


alpha = 0.05


print("Sample 1 mean:", np.mean(sample1))
print("Sample 2 mean:", np.mean(sample2))
print("T-statistic:", t_statistic)
print("P-value:", p_value)


if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the population means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the population means.")
    

"""
We first generate two random samples of equal size (sample1 and sample2) using the np.random.normal function.
   These samples are drawn from normal distributions with the same mean (5) and standard deviation (2) for simplicity.
   
We then perform a two-sample t-test using stats.ttest_ind. This function calculates the t-statistic and p-value for the two samples.

We define the significance level alpha (usually 0.05), which is the threshold for determining whether to reject the null hypothesis

We print out the sample means, t-statistic, and p-value.

Finally, we make a decision about whether to reject the null hypothesis based on the p-value. If the p-value is less than alpha,
  we reject the null hypothesis and conclude that there is a significant difference between the population means; otherwise,
  we fail to reject the null hypothesis.
  
  """
```

Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV
program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard
deviation of 1.5. Estimate the population mean with a 99% confidence interval.

To estimate the population mean with a 99% confidence interval, we can use the following formula:
    
   CI = x̄ ± z * (s / √n)

  where:
        CI is the confidence interval
        x̄ is the sample mean (4 ads)
        z is the critical value from the standard normal distribution for a 99% confidence interval, which is 2.576
        s is the sample standard deviation (1.5 ads)
        n is the sample size (50 viewers)
        
    Substituting the values from the problem, we get:
        
     CI = 4 ± 2.576 * (1.5 / √50) = 4 ± 0.47
    
    Therefore, the 99% confidence interval for the population mean number of ads watched by viewers during a TV program is 3.53 to 4.47 ads.
    
    This means that we are 99% confident that the true population mean number of ads watched by viewers during a TV program is
        between 3.53 and 4.47 ads.
        
    
    It is important to note that this is just an estimate of the population mean.
       The true population mean may be higher or lower than the confidence interval.
